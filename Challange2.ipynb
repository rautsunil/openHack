```python
def normalize(arr):
    arr = arr.astype('float')
    # Do not touch the alpha channel
    for i in range(3):
        minval = arr[...,i].min()
        maxval = arr[...,i].max()
        if minval != maxval:
            arr[...,i] -= minval
            arr[...,i] *= (255.0/(maxval-minval))
    return arr 
def normalize_intensity():
    for root, dirs, files in os.walk("gear_images", topdown=False):
        for name in files:
            path = root+"/"+name
            print(path)
            image = Image.open(path)
            arr = np.array(image)
            new_img = Image.fromarray(normalize(arr).astype('uint8'),'RGB')
            new_img.save("gear_images_normalized"+"/"+name)

```

```python
def equliize(path):
    image = Image.open(path)
    img_pil = Image.fromarray(np.uint8(image))
    img_pil = ImageOps.equalize(img_pil)
    plt.imshow(img_pil)
    return img_pil
```

```python
def platHistogram(path):
#     image = mpimg.imread(path)
#     lum_img = img[0,0,0]
#     plt.hist(lum_img.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
    image = cv.imread(path)
    chans = cv.split(image)
    colors = ("b", "g", "r")
    plt.figure()
    plt.xlabel("Bins")
    plt.ylabel("# of Pixels")
    features = []

    # loop over the image channels
    for (chan, color) in zip(chans, colors):
        # create a histogram for the current channel and
        # concatenate the resulting histograms for each
        # channel
        hist = cv.calcHist([chan], [0], None, [256], [0, 256])
        features.extend(hist)

        # plot the histogram
        plt.plot(hist, color = color)
        plt.xlim([0, 256])
        
    return hist
```

```python
platHistogram("gear_images_resized/884221.jpeg")
```

```python
def padding(infile):
    size = 128

    outfile = '1a.jpeg'
    outfile2 = '1b.jpeg'


    fill_color = (255, 255, 255, 255)

    img = Image.open(infile)

    img.thumbnail((size, size), Image.ANTIALIAS)

    x, y = img.size

    new_img = Image.new('RGB', (size, size), fill_color)
    new_img.paste(img, (int((size - x ) / 2), int((size - y) / 2)))


    new_img.save(outfile2, "JPEG")
    img.save(outfile, "JPEG")
    showImage(outfile)
```
